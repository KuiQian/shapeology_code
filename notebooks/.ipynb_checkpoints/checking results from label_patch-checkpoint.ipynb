```python
cell_size=ex.shape[0]
center=cell_size/2.
footprint=np.ones([cell_size,cell_size])>1
center_circle=np.ones([cell_size,cell_size])>1

for i in range(cell_size):
    for j in range(cell_size):
        footprint[i,j]=((i-center)**2+(j-center)**2)<=center**2
        center_circle[i,j]=((i-center)**2+(j-center)**2)<=(center**2)/2.
ring=footprint*1-center_circle*1

ring_pixels=ex[ring==1]
center_pixels=ex[center_circle==1]
np.mean(center_pixels),np.std(center_pixels),np.mean(ring_pixels),np.std(ring_pixels)
```

```python
center_circle_area=sum(center_circle.flatten())
ring_area=sum(ring.flatten())


center_circle_area,ring_area
```

```python
figure(figsize=(5,10))
subplot(121)
imshow(ring*ex,cmap='gray')
subplot(122)
imshow(center_circle*ex,cmap='gray');

ring_pixels=ex[ring==1]
center_pixels=ex[center_circle==1]
ring_pixels.shape,center_pixels.shape
```

```python
np.mean(center_pixels),np.std(center_pixels),np.mean(ring_pixels),np.std(ring_pixels)
```
