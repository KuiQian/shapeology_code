```python
filename='../data/MD657-F23-2017.02.17-23.13.17_MD657_2_0068_lossless.jp2'
!ls -lh $filename
```

```python
## Be prepared, loading a 2GB image takes ~15 minutes!
jp2 = glymur.Jp2k(filename)
fullres = jp2[:]
```

```python
x,y,z=fullres.shape
x,y,z
```

```python
cx=int(x/2)
cy=int(y/2)
d=500
cx,cy
```

```python
window=fullres[cx-d:cx+d,cy-d:cy+d,:]
window.shape
figure(figsize=[15,15])
imshow(window)
```

```python
import pickle
pickle.dump(window,open('../data/window.pkl','bw'))
```
