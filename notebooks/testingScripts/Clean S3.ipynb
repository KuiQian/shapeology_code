```python
dir = 's3://mousebraindata-open/MD657/'
_list=!aws s3 ls $dir
```

```python
L=!aws s3 ls $dir | grep tgz

import re
pat=re.compile(r'N1.*_(\d{4})_lossless')
IL=[]
for line in L:
    m=pat.search(line)
    if m:
        i=int(m.group(1))
        #print(line,i)
        IL.append((i,line))

SL=sorted(IL,key=lambda x:x[0])
for i,line in SL:
    print(line)
```
